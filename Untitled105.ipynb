### chi squared test

```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
```

```python
df=pd.read_csv("/content/global-data-on-sustainable-energy (1).csv")
df
```

```python
df_chi=pd.crosstab(df["Primary energy consumption per capita (kWh/person)"],df["Renewable energy share in the total final energy consumption (%)"])
```

```python
df_chi
```

```python
observedvalue=df_chi.values
```

```python
observedvalue
```

```python
value=stats.chi2_contingency(df_chi)
value
```

```python
expcted_value=value[3]
expcted_value
```

```python
datafreedomvalue =3434 * 2586
```

```python
criticalval=stats.chi2.ppf(q=0.95,df=8880324)
criticalval
```

```python
chisquare=sum([((o-e)**2)/e for o,e in zip(observedvalue,expcted_value)])
```

```python
chisquare
```

```python
chisqre=chisquare.sum()
```

```python
if chisqre >=criticalval:
  print("null hypothesis h0 is rejected,ha is accepted")
  print("there is some realationship between two categorical column")
else:
  print("null hypothesis h0 is accpted,ha is rejected")
  print("there is no realationship between two categorical column")
```
